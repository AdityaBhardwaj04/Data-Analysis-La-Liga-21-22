```python
# Import the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ll_df = pd.read_csv('C:\\Users\\KIIT\\Desktop\\La Liga\\LaLigaDataSet.csv', encoding='latin-1')
ll_df.head()
```

```python
ll_df.info()
```

```python
ll_df.describe()
```

```python
ll_df.isna().sum()
```

```python
ll_df['MinsPerMatch'] = (ll_df['Minutes Played'] / ll_df['Match Played']).astype(int)
ll_df['GoalsPerMatch'] = (ll_df['Goals'] / ll_df['Match Played']).astype(float)
print(ll_df)
```

```python
#Total Goals

Total_Goals = ll_df['Goals'].sum()
print(Total_Goals)
```

```python
#Penalty Goals

Total_PenaltyGoals = ll_df['Penalty Kicks'].sum()
print(Total_PenaltyGoals)
```

```python
#Penalty Attempts

Total_PenaltyAttempts = ll_df['Penalty Attempted'].sum()
print(Total_PenaltyAttempts)
```

```python
#Pie Chart for penalties missed vs scored

plt.figure(figsize=(13,6))
pl_not_scored = ll_df['Penalty Attempted'].sum() - Total_PenaltyGoals
data = [pl_not_scored, Total_PenaltyGoals]
labels = ['Penalties missed', 'Penalty Scored']
color = sns.color_palette('Set2')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
# Unique Positions

ll_df['Position'].unique()
```

```python
# Total FW Players

ll_df[ll_df['Position'] == 'FW']
```

```python
#Players from different nations

np.size((ll_df['Nationality'].unique()))
```

```python
# Most players come from which nation

nationality = ll_df.groupby('Nationality').size().sort_values(ascending=False)
nationality.head(10).plot(kind = 'bar', figsize=(12,6), color = sns.color_palette("magma"))
```

```python
#Clubs with maximum players in their squads

ll_df['Club'].value_counts().nlargest(5).plot(kind='bar', color=sns.color_palette('viridis'))
```

```python
#Clubs with least players in their squad

ll_df['Club'].value_counts().nsmallest(5).plot(kind='bar', color=sns.color_palette('viridis'))
```

```python
# Players based on age groups

Under20 = ll_df[ll_df['Age'] <= 20]
age20_25 = ll_df[(ll_df['Age'] > 20) & (ll_df['Age'] <= 25)]
age25_30 = ll_df[(ll_df['Age'] > 25) & (ll_df['Age'] <= 30)]
Above30 = ll_df[ll_df['Age'] > 30]
```

```python
x = np.array([Under20['Player'].count(), age20_25['Player'].count(),age25_30['Player'].count(), Above30['Player'].count()])
mylabels = ["<=20", ">20 & <=25", ">25 & <=30", ">30"]
plt.title('Total Players with Age Groups',fontsize = 20)
plt.pie(x,labels = mylabels, autopct="%.1f%%")
plt.show( )
```

```python
#Total under 20 players in each club

players_under_20 = ll_df[ll_df['Age'] < 20]
players_under_20['Club'].value_counts().plot(kind='bar', color=sns.color_palette("cubehelix"))
```

```python
#Under 20 players in Barcelona

players_under_20[players_under_20['Club'] == 'Barcelona']
```

```python
#Average age of players inn each club

plt.figure(figsize = (12,6))
sns.boxplot(x = 'Club', y = 'Age',data = ll_df)
plt.xticks(rotation = 90)
```

```python
num_player = ll_df.groupby('Club').size()
data = (ll_df.groupby('Club')['Age'].sum()) / num_player
data.sort_values(ascending = False)
```

```python
# Total assists from each club
Assits_by_clubs= pd.DataFrame(ll_df.groupby('Club', as_index=False)['Assist'].sum())
sns.set_theme(style="whitegrid",color_codes=True)
ax=sns.barplot(x='Club' , y= 'Assist' ,data=Assits_by_clubs.sort_values(by='Assist'),palette='Set2')
ax.set_xlabel("Club", fontsize=30)
ax.set_ylabel("Assists" , fontsize=20)
plt.xticks(rotation = 75)
plt.rcParams["figure.figsize"] = (20, 8)
plt.title('Plot of Clubs vs Total Assists', fontsize = 20)
```

```python
#Top 10 Assists

top_10_assists = ll_df[['Player', 'Club', 'Assist', 'Match Played']].nlargest(n = 10, columns = 'Assist')
top_10_assists
```

```python
Goals_by_clubs=pd.DataFrame(ll_df.groupby('Club',as_index=False)['Goals'].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax=sns.barplot(x = 'Club', y = 'Goals', data = Goals_by_clubs.sort_values(by='Goals'), palette='rocket')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Goals",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"] = (20,8)
plt.title('Plot of Clubs vs Total Goals',fontsize= 20)
```

```python
#Most goals by players

top_10_goals = ll_df[['Player', 'Club', 'Goals', 'Match Played']].nlargest(n=10, columns = 'Goals')
top_10_goals
```

```python
#Goal per match

top_l0_goals_per_match = ll_df[['Player', 'GoalsPerMatch', 'Match Played', 'Goals']].nlargest(n = 10, columns = 'GoalsPerMatch')
top_l0_goals_per_match
```

```python
# Pie Chart - Goals with assist and without assist

plt.figure(figsize=(14,7))
assists = ll_df['Assist'].sum()
data = [Total_Goals-assists , assists ]
labels = ['Goals without assists' , 'Goals with assists']
color = sns.color_palette('Set1')
plt. pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt. show()
```

```python
# Top 10 players with most Yellow Card

ll_yellow = ll_df.sort_values(by = 'Yellow Card', ascending = False)[:10]
plt.figure(figsize = (20,6))
plt.title("Players with most Yellow Card")
c = sns.barplot(x = ll_yellow['Player'], y = ll_yellow['Yellow Card'], label = 'Players', color = 'yellow')
plt.ylabel("Number of Yellow Card")
c.set_xticklabels(c.get_xticklabels(), rotation = 45)
c
```
